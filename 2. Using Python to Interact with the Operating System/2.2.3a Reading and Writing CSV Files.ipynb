## Reading CSV Files
* Parsing the file using CSV module and unpacking the value from CSV file

```python
import csv
f = open("csv_file.txt")
csv_f = csv.reader(f)    # Parses the file using the CSV module
```

```python
for row in csv_f: 
    name, phone, role = row   # unpakcs the value from the CSV file
    print(f"{name}, {phone}, {role}")
f.close()
```

## Generating CSV
* .writerow() method writes one row at a time
* .writerows() method writes all of them together


**No output but csv file will be created in the same folder as where we open this .ipynb file**

```python
hosts = [["workstation.local", "192.168.25.46"], ["webserver.cloud", "10.2.5.6"]]

with open("hosts.csv", "w") as hosts_csv:
    writer = csv.writer(hosts_csv)
    writer.writerows(hosts)
```
